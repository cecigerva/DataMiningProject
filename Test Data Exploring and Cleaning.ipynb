```python
bins = [0, 25, 45, 65, 90]
census_test_sub['age_dis'] = pd.cut(census_test_sub['age'], bins, labels = ['young', 'adult', 'senior', 'old'], include_lowest = True)

print(census_test_sub['age'])
print(census_test_sub['age_dis'])
```

```python
age = pd.get_dummies(census_test_sub['age_dis'])
age
```

```python
census_test_sub = pd.concat([census_test_sub.drop(['age_dis'],axis=1), age],axis=1)
```

```python
bins = [0, 39, 40, 70]
census_test_sub['hr_per_week_dis'] = pd.cut(census_test_sub['hr_per_week'], bins, labels = ['part=time', 'full-time', 'over-time'], include_lowest = True)

print(census_test_sub['hr_per_week'])
print(census_test_sub['hr_per_week_dis'])
```

```python
hr_per_week = pd.get_dummies(census_test_sub['hr_per_week_dis'])
hr_per_week
```

```python
census_test_sub = pd.concat([census_test_sub.drop(['hr_per_week_dis'],axis=1), hr_per_week],axis=1)
```
